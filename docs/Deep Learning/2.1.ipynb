# 2.1数据操作summary

## 张量的创建与基础属性

- 用torch.arange(n)创建包含连续元素的张量（如torch.arange(12)生成 0-11 的张量）；
- 用.shape查看张量的形状（如 1 维张量形状为torch.Size([12])）；
- 用.numel()查看张量的元素总数（如 12 个元素的张量，结果为 12）。

## 张量的形状重塑

通过.reshape(shape)调整张量维度（如把 1 维 12 元素的张量，重塑为3×4的 2 维张量）

## 张量的创建方式

- 可通过torch.zeros/ones/rand创建全 0、全 1、随机值张量；
- 用torch.tensor()手动传入数据（如torch.tensor([[2,1,4,3], ...])）创建自定义张量。

## 张量的基本运算

加减乘除（+、-、*、/）、幂运算（$ ** ）等，运算后得到同形状的张量。

## 张量的拼接与求和

- 用torch.cat(tensors, dim)按指定维度拼接张量（dim=0为行拼接，dim=1为列拼接）；
- 用.sum()对张量所有元素求和（如示例中张量求和结果为66.）。

## 张量的广播机制

当两个张量形状不完全匹配但符合 “广播规则” 时，会自动扩展形状后执行逐元素运算：

e.g. a(3,1)和b(1,2)相加，自动扩展为(3,2)形状后运算，得到3×2的张量。

## 张量的索引与切片操作

- 索引：通过下标取部分元素（如X[-1]取最后一行，X[1:3]取第 2-3 行）；
- 赋值：支持单个元素修改（X[1,2]=9）、范围元素批量修改（X[0:2,:]=12修改前两行所有元素）。

## 张量的内存与原地操作

- 普通运算（如Y = Y+X）会创建新张量，内存地址（id()）改变；
- 原地操作（如Z[:] = X+Y、X += Y）会直接修改原张量，内存地址保持不变。

## 张量与 NumPy 的转换

用.numpy()将 PyTorch 张量转为 NumPy 数组；用torch.tensor(NumPy数组)将 NumPy 数组转回 PyTorch 张量；两者类型分别为numpy.ndarray和torch.Tensor。


## 单个元素张量的取值

对仅含 1 个元素的张量，可通过：
- .item()提取为 Python 标量；
- float()/int()直接转换为对应 Python 数值类型。
